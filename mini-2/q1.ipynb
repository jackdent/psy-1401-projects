```python
%matplotlib inline

import matplotlib.pyplot as plt
import random
```

```python
def pearce_hall((vn, an, rn), k, n):
    dn = rn - vn
    a_nplus1 = n * abs(dn) + (1 - n) * an
    v_nplus1 = vn + k * an * dn
    return (a_nplus1, v_nplus1)
```

```python
v0 = .0
a0 = .926
n = .166
k = .857
shock_rate = .333

ans, f1_vns, f2_vns = [a0], [v0], [v0]




#learn phase: the wi correspond to presentation
#recall phase: the wi correspond to recall? 

# Learning phase
for x in range(100):
    if random.random() > .5:
        # Show face 1
        if random.random() < shock_rate:
            an, f1_vn = pearce_hall((f1_vns[-1], ans[-1], 1), k, n)
            ans.append(an)
            f1_vns.append(f1_vn)
        else:
            an, f1_vn = pearce_hall((f1_vns[-1], ans[-1], 0), k, n)
            ans.append(an)
            f1_vns.append(f1_vn)
    else:
        # Show face 2
        an, f2_vn = pearce_hall((f2_vns[-1], ans[-1], 0), k, n)
        ans.append(an)
        f2_vns.append(f2_vn)
        
# Reversal phase
for x in range(100):
    if random.random() > .5:
        # Show face 2
        if random.random() < shock_rate:
            an, f2_vn = pearce_hall((f2_vns[-1], ans[-1], 1), k, n)
            ans.append(an)
            f2_vns.append(f2_vn)
        else:
            an, f2_vn = pearce_hall((f2_vns[-1], ans[-1], 0), k, n)
            ans.append(an)
            f2_vns.append(f2_vn)
    else:
        # Show face 1
        an, f1_vn = pearce_hall((f1_vns[-1], ans[-1], 0), k, n)
        ans.append(an)
        f1_vns.append(f1_vn)
```
